```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv('sonar.csv', header = None)
```

```python
X = dataset.iloc[:, 0:60].values
y = dataset.iloc[:, 60].values
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
#feature encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
y
```

```python
#splitting into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Building ANN
#Importing Keras Libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
# Initialize ANN
model = Sequential()
```

```python
#Build input Layer & Hidden Layer with droupout
model.add(Dense (units=30, activation = 'relu', kernel_initializer = 'uniform', input_dim=60)) #dim gives the number of nodes of input
model.add(Dropout(rate=0.1))
#units give no. of hidden layers
#hidden layers are approximately half to that of input layers
```

```python
#Adding Second Hidden Layer
model.add(Dense (units=30, activation = 'relu', kernel_initializer = 'uniform'))
model.add(Dropout(rate=0.1)) #drops 10% of the nodes
```

```python
#Adding output Layer
model.add(Dense(units=1, activation = 'sigmoid',  kernel_initializer = 'uniform'))
```

```python
#Compiling our ANN Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
#fitting ANN to training set
model.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
#Predicting on test set
y_pred = model.predict(X_test) > 0.5
```

```python
#evaluating using confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
